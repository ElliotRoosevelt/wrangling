```python
shark['Type'].value_counts() #to see what the categories are
```

```python
typetemp=shark['Type'] #making the Type column in shark df into a new variable temp.
#combining the values into watercraft or NA categories
type=typetemp.replace(['Sea Disaster', 'Boat', 'Boating', 'Boatomg'], 'Watercraft')
type.value_counts()
```

```python
#3
#I used this link http://www.vcsc.virginia.gov/pretrialdataproject2017.html to find the csv file because I did not know how to use the pdf file in github
VA=pd.read_csv('/content/October 2017 Cohort_Virginia Pretrial Data Project_Deidentified FINAL Update_10272021.csv', low_memory=False)
```

```python
VA.head()
#VA.columns.tolist()
```

```python
#temp for the Whether Column separate from df
released=VA['WhetherDefendantWasReleasedPretrial']
print(released.unique(),'\n')
print(released.value_counts(),'\n')
```

```python
 # In the codebook, the 9's are "unclear"
release = released.replace(9,np.nan)
print(release.value_counts(),'\n')
```

```python
sum(release.isnull()) # 31 missing values
```

```python
#replacing release into Whether column in df
VA['WhetherDefendantWasReleasedPretrial'] = release
```

```python
del release
```

```python
#4 clean imposed
length = VA['ImposedSentenceAllChargeInContactEvent']
type= VA['SentenceTypeAllChargesAtConvictionInContactEvent']
length.shape
```

```python
print( length.unique()  , '\n') #can see there are ' ' values
```

```python
length = pd.to_numeric(length,errors='coerce') # Coerce length to numeric
length_NA =length.isnull() # Create a missing dummy
```

```python
print( np.sum(length_NA),'\n') #22900 to 9000 values significant loss
```

```python
print( pd.crosstab(length_NA, type), '\n') # Category 4 is cases where the charges were dismissed
```

```python
length = length.mask( type == 4, 0) # Replace length with 0 when type ==4
length = length.mask( type == 9, np.nan) # Replace length with np.nan when type == 9
```

```python
length_NA = length.isnull() # Create a new missing dummy
print( pd.crosstab(length_NA, type), '\n')
print( np.sum(length_NA),'\n') # 274 missing but 9000 is now not missing
```

```python
VA['ImposedSentenceAllChargeInContactEvent'] = length # Replace data with cleaned version
del length, type # Delete temporary length/type variables
```
